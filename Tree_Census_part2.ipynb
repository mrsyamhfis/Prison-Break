#Data cleaning

Dealing with missing values

```
tc_subset['health'].value_counts()
```

```
mask = ((tc_subset['status']=='Stump')|(tc_subset['status']=='Dead'))
```

```
tc_subset.loc[mask]=tc_subset.loc[mask].fillna('Not Applicable')
```

```
tc_subset[tc_subset['status']=='Stump']
```

```
tc_subset.isna().sum()
```

```
tc_subset['problems'].fillna('None', inplace = True)
tc_subset['health'].fillna('Good', inplace = True)
tc_subset['spc_latin'].fillna('No Observation', inplace = True)
tc_subset['sidewalk'].fillna('No Damage', inplace = True)
```

Dealing with outliers

```
big_trees = tc_subset[(tc_subset['tree_dbh']>60)|(tc_subset['stump_diam']>60)]
big_trees
```

```
tc_subset=tc_subset[(tc_subset['tree_dbh']<=60)|(tc_subset['stump_diam']<=60)]
tc_subset
```

```
tc_subset_alive = tc_subset[tc_subset['status']=='Alive']
tc_subset_dead_or_stump = tc_subset[(tc_subset['status']=='Dead')|(tc_subset['status']=='Stump')]
```

```
tc_subset_alive.groupby('spc_latin')['tree_dbh'].describe()
```

```
stats_alive = tc_subset_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin','25%','75%']]
stats_alive
```

```
tc_subset_alive = tc_subset_alive.merge(stats_alive, on='spc_latin', how='left')
```
